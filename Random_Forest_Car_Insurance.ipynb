```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('/content/Cost of car insurance by state.xlsx')
```

```python
data.shape
```

```python
data.describe()
```

```python
data.head()
```

```python
# creating a rank column and passing the returned rank series
data["Rank"] = data["State"].rank()
```

```python
data.drop(columns='Rank', axis=1)
```

```python
# sorting w.r.t name column
data.sort_values("State", inplace = True)
```

```python
df = sns.heatmap(data.corr(), annot=True)
```

```python
X = data.drop('State', axis=1)
```

```python
y=data['State']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X,y)
```

```python
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import *
print("Classification report - \n", classification_report(y_test,y_pred))
```
